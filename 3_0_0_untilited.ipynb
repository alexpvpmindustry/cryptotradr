```python
coli_with_na = np.where(np.sum(pd.isna(data["Close"]))>0)[0]
for coli in coli_with_na:
    print(columns[coli])
```

```python
opens= data["Open"].drop(columns=[columns[coli] for coli in coli_with_na])
closes = data["Close"].drop(columns=[columns[coli] for coli in coli_with_na])
# Open	High	Low	Close	Volume
highs = data["High"].drop(columns=[columns[coli] for coli in coli_with_na])
lows = data["Low"].drop(columns=[columns[coli] for coli in coli_with_na])
volumes = data["Volume"].drop(columns=[columns[coli] for coli in coli_with_na])
```

```python
opens.shape,closes.shape,highs.shape,lows.shape,volumes.shape
```

```python
import pandas_ta as ta
```

```python
%%time
training_X=[]
training_Y=[]

check_correlation_Y = []
ticker_symbols = []
for ticker in opens.columns[:]:
    temp_df_full = pd.DataFrame(data={"Open":opens[ticker],
                                 "High":highs[ticker],
                                 "Low":lows[ticker],
                                 "Close":closes[ticker],
                                 "Volume":volumes[ticker],
                                })
    t=0
    while True:
        try:
            temp_df_full.iloc[34+t+1]
        except IndexError :
            break
        temp_df = temp_df_full.iloc[0+t:34+t]
        temp_normed = pd.concat([temp_df["Open"]/(temp_df["Open"].iloc[0]),
                          temp_df["High"]/(temp_df["High"].iloc[0]),
                          temp_df["Low"]/(temp_df["Low"].iloc[0]),
                          temp_df["Close"]/(temp_df["Close"].iloc[0]),
                          temp_df["Volume"]/(temp_df["Volume"].iloc[0])],axis=1)
        temp_ = pd.concat([temp_df["Open"],
                          temp_df["High"],
                          temp_df["Low"],
                          temp_df["Close"],
                          temp_df["Volume"]],axis=1)
        #temp_df.loc["Open"] = temp_
        temp_df = temp_.copy()
        ta_df_temp = pd.concat([
            temp_df.ta.sma()/(temp_df["Open"].iloc[0]),
            temp_df.ta.ema()/(temp_df["Open"].iloc[0]),
            temp_df.ta.donchian()/(temp_df["Open"].iloc[0]),
            temp_df.ta.rsi(),
            #temp_df.ta.ichimoku()[0],
            temp_df.ta.fisher(),
            temp_df.ta.atr()/(temp_df["Open"].iloc[0]),
            temp_df.ta.squeeze(),
            temp_df.ta.bbands()/(temp_df["Open"].iloc[0]),
            temp_df.ta.macd(),
            temp_df.ta.nvi()
            ],axis=1)
        ans = ta_df_temp.iloc[33].values
        final_df = temp_df_full.iloc[34+t+1]
        result = (final_df["Close"]-final_df["Open"])/(final_df["Close"])
        training_X.append(ans)
        training_Y.append(result)
        check_correlation_Y.append( temp_df["Open"].iloc[0] )
        t+=10
        ticker_symbols.append(ticker)
    print(ticker,end=" ")
    #print(temp_df.tail())
    #print(temp_df.ta.nvi().tail())
    
    #print(np.sum(pd.isna(ta_df_temp).values), result)
```

```python
training_X =np.asarray(training_X)
check_correlation_Y = np.asarray(check_correlation_Y)
training_X.shape,check_correlation_Y.shape
```

```python
fig,axx= plt.subplots(5,5,figsize=(10,10))
axx=axx.ravel() 
for y_scatter_i,ax in zip(range(len( training_X[0] )),axx):
    ax.scatter(check_correlation_Y,training_X[:,y_scatter_i])
    ax.set_title( y_scatter_i)
plt.suptitle("correlation between indicators and open price")
plt.tight_layout()
```
